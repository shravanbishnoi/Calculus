```python
# Q-Solve using Euler method dx/dt = x
import plotly.express as px
import plotly.graph_objects as go
import math
```

```python
# Using EULER METHOD for ploting dx/dt = math.exp(t)
lst_t = [] # initializer for a list of value on xaxis
lst_x = [] # initializer for a list of value on yaxis
t = 0.0
while t<5:
  lst_t.append(t)
  x = math.exp(t)
  lst_x.append(x)
  t += 0.1
```

```python
# Using EULER METHOD for ploting dx/dt = math.exp(t) without computing mathematically
lst_t1 = [] # initializer for a list of value on xaxis
lst_x1 = [] # initializer for a list of value on yaxis
delta_t = 0.01  # Given delta
x_hat = 1  # Given x_hat
t1 = 0
while t1<5:
  lst_t1.append(t1)
  # Calculating x for given delta
  x_hat += (x_hat)*delta_t
  lst_x1.append(x_hat)
  t1 += 0.01 # increment
```

```python
# for comparing we plot both on the same page
fig = go.Figure()
fig.add_trace(go.Scatter( name = 'Analytical',x = lst_t, y=lst_x))

fig.add_trace(go.Scatter(name = 'Numerical', x = lst_t1, y= lst_x1))

fig.update_layout(barmode='group', xaxis_title='Exponential')
fig.show()
```
