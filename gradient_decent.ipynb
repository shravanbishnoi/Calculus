Differentiation:
(i) Numerical (Approx answer)
(ii) Symbolic (Exact answer)
(iii) Using Automatic(autodiff) method (Exact answer)

```python
import autograd.numpy as np
from autograd import grad
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

To find derivative for one variable

```python
def my_function(x):
  return np.sin(x)
gradient_func = grad(my_function)
gradient = gradient_func(2.0)
```

Finding and plotting sin(x) using Numerical, Symbolic and Automatic and comparing all:-
(i) Using Numerical- 
$$f\prime(x)=\frac {sin(x + Δx) - sin(x)} {Δx}$$

```python
delta_x = 0.5
x = 0 
lst_x = []
lst_y = []
while x<8*np.pi: # taking 4 cycles
  lst_x.append(x)
  y = (np.sin(x + delta_x) - np.sin(x))/delta_x
  lst_y.append(y)
  x += delta_x
```

(ii) Using Symbolic-

```python
x_s = 0
lst_x_s = []
lst_y_s = []
while x_s<8*np.pi:
  lst_x_s.append(x_s)
  y = np.cos(x_s)
  lst_y_s.append(y)
  x_s+= delta_x
```

(iii) Using Automatic-

```python
delta_x = 0.5
x_a = 0.0
lst_x_a = []
lst_y_a = []
while x_a<8*np.pi:
  lst_x_a.append(x_a)
  lst_y_a.append(gradient_func(x_a))
  x_a += delta_x
```

Plotting for all three

```python
# Plottings comparison for the three methods- Numerical, Symbolic and automatic
plt.plot(lst_x, lst_y, label = 'Numerical')
plt.plot(lst_x_s, lst_y_s, label = 'Symblic')
plt.plot(lst_x_a, lst_y_a, label = 'Automatic')
plt.legend()
plt.show()
```

Gradient Descent:
Calculating Local $minima$ of the function $f(x)=sin(x)$ using Automatic method

```python
eta = 0.01
xn = 1.0 # A value to start from
var = True # loop variable
while var:
  if gradient_func(xn)>0:
    xn -= 0.1 # In case of minima if slope is +ve we move left from the starting point
    xn1 = xn - eta*(gradient_func(xn))
    # xN = xn - eta*(gradient_func(xn))
    # xn += xN
    if gradient_func(xn1)<0:
      print(xn1/np.pi) # prints the minima in radian
      var = False
  elif gradient_func(xn)<0:
    xn += 0.1 # In case of minima if slope is -ve we move right from the starting point
    xn1 = xn - eta*(gradient_func(xn))
    if gradient_func(xn1)>0:
      print(xn1/np.pi) # prints the minima in radian
      var = False
```

Gradient Descent:
Calculating Local $minima$ of the function $f(x)=x^2 + y^2$ using Automatic method

Below function returns the gradient of $f(x)=x^2+y^2$

```python
def myfunc(X):
  x,y = X
  return x**3 + y**2
gradient_func = grad(myfunc)
```

```python
eta = 0.01
xn = [15.87,78.98] # values to start from
previous = [0.0,0.0]   # accumulator to keep previous values track
var = True
while var:
  xn[0] = xn[0] - eta*(gradient_func(xn)[0]) # gradient decent for first
  xn[1] = xn[1] - eta*(gradient_func(xn)[1]) # gradient decent for second
  
  if abs(gradient_func(xn)[0] - gradient_func(previous)[0])<0.0001 and abs(gradient_func(xn)[1] - gradient_func(previous)[1])<0.0001:
    var = False
    print(xn)
  previous = xn.copy() # avoiding loss of values
```
