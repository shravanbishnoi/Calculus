```python
import sympy as sp
import math
from scipy.integrate import quad
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def f(x):
  return ((a**2*(math.cos(x))**2) + (b**2*(math.sin(x))**2))**(0.5)
```

```python
a = 2.0 # a ranging from 2 to 4
x_list = [] # a/b for x-axis
lst_a = [] # list of values of a
lst_b = [] # list of values of b
lst_f = [] # values after integrating the function taking a and b's combinations

while a<=4:
  b = 1.0 # setting b = 1 as constant, when we take combinations of a and b then second while loop will be run. and indent rest of the code
  while b<=2.0:
    # print(a,b)
    result = quad(f,0,math.pi*2)
    lst_f.append(result[0])
    lst_a.append(a)
    lst_b.append(b)

    # if a==b:
    #   print(result[0])
    b += 0.1
  a += 0.1
print(len(lst_a))
```

```python
# Create figure and axes
fig = plt.figure()
ax = plt.axes(projection='3d')

# Plot data
ax.plot3D(lst_a, lst_b, lst_f, 'red')

# Add labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot')

# ShowÂ plot
plt.show()
```

```python
a = 2.0 # a ranging from 2 to 4
b = 1.0 # setting b = 1 as constant, when we take combinations of a and b then second while loop will be run. and indent rest of the code

x_list = [] # a/b for x-axis
lst_a = [] # list of values of a
lst_b = [] # list of values of b
lst_f = [] # values after integrating the function taking a and b's combinations
lst_A = [] # values yield by first formula math.pi*(((6*a)/5) + ((3*b)/4))
lst_B = [] # values yield by second formula  math.pi*2*(((a**2+b**2)/2)**(0.5))
lst_C = [] # values yield by third formula math.pi*2*((a+b)/2)
lst_R = [] # values yield by Ramanujan formula math.pi*((3*(a+b))- ((3*a+b)*(a+3*b))**(0.5))
while a<=4:
  result = quad(f,0,math.pi*2)
  lst_f.append(result[0])
  lst_a.append(a)
  lst_b.append(b)
  A = math.pi*(((6*a)/5) + ((3*b)/4))
  lst_A.append(A)
  B = math.pi*2*(((a**2+b**2)/2)**(0.5))
  lst_B.append(B)
  C = math.pi*2*((a+b)/2)
  lst_C.append(C)
  x_list.append(a/b)
  R = math.pi*((3*(a+b))- ((3*a+b)*(a+3*b))**(0.5))
  lst_R.append(R)
  a += 0.1

# print(lst_f)
# print(len(lst_f))
# print(lst_a)
# print(len(lst_a))
# print(lst_b)
# print(len(lst_a))
```

```python
plt.plot(x_list, lst_A, label = 'difference of f and A')
plt.plot(x_list, lst_B, label = 'difference of f and B')
plt.plot(x_list, lst_C, label = 'difference of f and C')
plt.plot(x_list, lst_R, label = "Ramanujan's")

plt.legend(title = 'Comparison graph with different formulae:')
plt.show()
```

```python
dif_A = [] # dif for first formula
dif_B = [] # dif for second formula
dif_C = [] # dif for third formula
dif_R = [] # dif for Ramanujan's formula
for i in range(len(x_list)):
  # calculating difference with respect to function values and taking absolute percentage by diving it with expected values
  dif_A.append((abs((lst_f[i]-lst_A[i])/lst_f[i]))*100)
  dif_B.append((abs((lst_f[i]-lst_B[i])/lst_f[i]))*100)
  dif_C.append((abs((lst_f[i]-lst_C[i])/lst_f[i]))*100)
  dif_R.append((abs((lst_f[i]-lst_R[i])/lst_f[i]))*100)
```

```python
plt.plot(x_list, dif_A, label = 'difference of f and A')
plt.plot(x_list, dif_B, label = 'difference of f and B')
plt.plot(x_list, dif_C, label = 'difference of f and C')
plt.plot(x_list, dif_R, label = "Ramanujan's")

plt.legend(title = 'Difference graph comparison with respect to different formulae in percentage error:')
plt.show()
```
